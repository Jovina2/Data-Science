```python
import matplotlib.pyplot as plt
import numpy as np 
#plot 1
plt.suptitle("GRAPH")
x=np.array([1,2])
y=np.array([2,4])
plt.subplot(2,3,1)
plt.plot(x,y,'o')
plt.title("Title 1")
plt.xlabel("x label",color='blue')
plt.ylabel("y label",color='blue')

#plot 2
z=np.array([3,8,1,10])
w=np.array([6,2,7,11])
plt.subplot(2,3,2)
plt.plot(z,marker='*',color='orange',linewidth='2.5')
plt.plot(w,color='blue',linestyle = '--', linewidth = 2.5)

#plot 3
x1=np.array([3,8,1,10])
plt.subplot(2,3,3)
plt.plot(x1,marker='o',mfc='yellow',mec='black',color='blue')

#plot 4
x1=np.array([3,6])
y1=np.array([1,2])
plt.subplot(2,3,4)
plt.plot(x1,y1)
plt.title("Title 2")
plt.grid()

#plot 5
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.subplot(2,3,5)
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x, y,s=sizes,alpha=0.5,c=colors)

#plot6
x = np.array(["A", "B", "C"])
y = np.array([4, 2, 6])
plt.subplot(2,3,6)
plt.bar(x, y,color='red',width = 0.1)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 20, 20, 25])
mylabels = ["35", "20", "20", "25"]
myexplode = [0.1, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
plt.pie(y,labels = mylabels, startangle = 90,explode=myexplode,shadow=True,colors=mycolors)
plt.legend()
plt.show() 
```
